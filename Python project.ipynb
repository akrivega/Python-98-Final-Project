```python
#importing packages that we will use
import numpy as np
import math
import matplotlib.pyplot as plt
import astropy
from astropy import units as u
from astropy import constants as const
from astropy.io import fits
#import lmfit
#from lmfit import minimize, Parameters, report_fit, fit_report
from IPython.display import Image
```

```python
#Open the file
apf_file = fits.open("ucb-amp194.fits.txt")
wav = fits.open('apf_wav.fits')
spec = fits.open('ramp.193.fits')
```

```python
spec = spec[0].data
wavelength = wav[0].data
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(np.append(wavelength[27,:], wavelength[28,:]), \
        np.append(spec[27, 1:], spec[28,1:]))
ax.set_ylim(0,2500)
```

```python
#check header
header = apf_file[0].header
print(header)
```

```python
#pull out data
image = apf_file[0].data
plt.plot(np.sum(image, axis=0)[500:1000])
```

```python
image_rot = np.rot90(image)
plt.figure(figsize=(20,20))
plt.imshow(image_rot, vmin = np.median(image_rot), vmax = np.median(image_rot)*1.1, origin = 'lower',cmap="gray")
plt.title('2D array data w/ contrast')
```
